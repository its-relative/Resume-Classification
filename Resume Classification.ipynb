```python
# prompt: import the above csv file

import pandas as pd

# Read the uploaded CSV file into a pandas DataFrame
df = pd.read_csv('all_resumes.csv')

# Print the DataFrame
print(df)
```

```python
import nltk
nltk.download('all')
```

```python
# prompt: apply the tokenization using nltk library for above data

from nltk.tokenize import word_tokenize

df['Tokens'] = df['Resume'].apply(word_tokenize)

# Print the DataFrame with the new column
print(df)
```

```python
import string as s
print(s.punctuation)
```

```python
# prompt: remove the punctuations from the above tokens

df['Tokens'] = df['Tokens'].apply(lambda x: [word for word in x if word not in s.punctuation])

# Print the DataFrame with the new column
print(df)
```

```python
# prompt: do normilization to the above data

df['Tokens'] = df['Tokens'].apply(lambda x: [word.lower() for word in x])

# Print the DataFrame with the new column
print(df)
```

```python
# prompt:  print the stop words list from nltk

nltk.download('stopwords')
stop_words = nltk.corpus.stopwords.words('english')
print(stop_words)
```

```python
# prompt: remove the stopwords from tokens_without_punctuation_lower

df['Tokens_without_stopwords'] = df['Tokens'].apply(lambda x: [word for word in x if word not in stop_words])

# Print the DataFrame with the new column
print(df)
```

```python
# prompt: count the frequency of each word from above tokens_without_punctuation_lower_without_stopwords and construct a bargraph on it

import matplotlib.pyplot as plt

# Create a dictionary to store the frequency of each word
word_frequency = {}

# Loop through each token in the DataFrame
for tokens in df['Tokens_without_stopwords']:
    for token in tokens:
        # If the word is not already in the dictionary, add it with a frequency of 1
        if token not in word_frequency:
            word_frequency[token] = 1
        # Otherwise, increase the frequency by 1
        else:
            word_frequency[token] += 1

# Sort the dictionary by frequency in descending order
sorted_word_frequency = sorted(word_frequency.items(), key=lambda item: item[1], reverse=True)

# Extract the top 20 most frequent words
top_20_words = sorted_word_frequency[:20]

# Extract the words and frequencies
words, frequencies = zip(*top_20_words)

# Create a bar chart
plt.bar(words, frequencies)

# Set the title and axis labels
plt.title('Frequency of Top 20 Words')
plt.xlabel('Words')
plt.ylabel('Frequency')

# Rotate the x-axis labels for better readability
plt.xticks(rotation=45)

# Show the plot
plt.show()
```

```python
# prompt:  print the unique words from the sorted_tokens save it in another variable

unique_words = []
for word, frequency in sorted_word_frequency:
    unique_words.append(word)

print(unique_words)
```

```python
# prompt: apply the lemmatization to the above data

nltk.download('wordnet')
from nltk.stem import WordNetLemmatizer

lemmatizer = WordNetLemmatizer()

df['Lemmatized_Tokens'] = df['Tokens_without_stopwords'].apply(lambda x: [lemmatizer.lemmatize(word) for word in x])

# Print the DataFrame with the new column
print(df)
```

```python
# prompt: give me the code for word cloud

from wordcloud import WordCloud

# Join all the tokens in the DataFrame into a single string
text = " ".join(df['Lemmatized_Tokens'].sum())

# Create a WordCloud object
wordcloud = WordCloud(width=800, height=600, background_color="white").generate(text)

# Plot the wordcloud
plt.figure(figsize=(10, 7))
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.show()
```
